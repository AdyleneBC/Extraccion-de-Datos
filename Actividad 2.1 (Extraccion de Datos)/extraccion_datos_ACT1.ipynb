**Actividad 2.1 (Extracción de Datos)**

```python
#Import the libraries to work with data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load from a .xlsx file without index
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

**_FILTROS BASICOS PARA VARIABLES CUALITATIVAS_**

```python
#Filter by object
filterObject = df[(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filterObject

#Descripcion: filtramos la variable-columna NOMBRE_VENDEDOR y devuelve solo las filas 
#que son TRUE respecto a LETICIA RAMIREZ HERNANDEZ
```

```python
#Filter by rows
#filter2 = df.iloc[3:10,:]  #Filas consecutivas
filterByRows = df.iloc[[2,5] , :]  #Filas no consecutivas
filterByRows

#Decripcion: Filtramos las filas 2 y 5, pues usamos el codigo de filas no consecutivas 
```

```python
#Filter by columns
filterByColumns = df.iloc[ : , 1:3] #Columnas consecutivas
#filter3 = df.iloc[ : , [1,4,6]] #Columnas no consecutivas
filterByColumns

#Descripcion: Filtramos las columnas consecutivas de la primera a la tercera (INDEX, FECHA_DOC, NOMBRE_VENDEDOR)
```

```python
#Carga desde un archivo .xlsx con indice
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=0)
df1
```

```python
#Paso 2: Aplicamos filtro de filas vs columnas
filterRowsVSColumns = df1.loc[["F39834", "F37043", "F37041"], ["NOMBRE_CLIENTE", "MARGEN_TOTAL_CALCULADO"]]
filterRowsVSColumns

#Descripcion: Filtramos las filas que tengan "F39834", "F37043", "F37041" contra las columnas
#que se llamen "NOMBRE_CLIENTE", "MARGEN_TOTAL_CALCULADO"
```

```python
#Filter 5
filterHEAD = df.head(10)  #muestra los primeros 10 registros
filterHEAD

#Descripcion: Filtramos los primeros 10 registros tomando en cuenta las filas
```

```python
#Filtro 6
filterTAIL = df.tail(2)  #muestra los 5 ultimos registros
filterTAIL

#Description: Filtramos los ultimos dos registros 
```

**_FILTROS BASICOS PARA VARIABLES CUANTITATIVAS_**


```python
#Filter by comparison
filterByComparison = df[df["MARGEN_TOTAL_CALCULADO"] > 7000]
filterByComparison

#Descripcion: Filtramos la columna MARGEN_TOTAL_CALCULADO que tenga registros mayores a 7000
```

**_FILTROS BASICOS CON CONETORES LOGICOS_**

```python
#Filter AND
filterAND = df[(df["PRECIO_UNITARIO"] < 2500) & (df["CVE_ART"] == "BOBINA001")]
filterAND

#Description: Filtramos la columna PRECIO_UNITARIO que tenga registros menores a 2500 y al mismo tiempo
# filtramos la columna CVE_ART que tenga registros igual a BOBINA001
```

```python
#Filter OR
filterOR = df[(df["NOMBRE_VENDEDOR"] == "DIEGO ARMANDO LEON DURAN") | (df["CVE_ART"] == "BOBINA001")]
filterOR

#Description: Filtramos la columna NOMBRE_VENDEDOR que tenga registros iguales a DIEGO ARMANDO LEON DURAN o
# filtramos la columna CVE_ART que tenga registros igual a BOBINA001
```

```python
#Filter NOT
filterNOT = df[~(df["CVE_DOC"] == "F39834")]
filterNOT

#Description: filtramos la columna CVE_DOC que no tenga el registro igual a F39834
```

```python
#Convertir archivo filtrado a CSV
filterObject.to_csv("filterByObject.csv")
filterByRows.to_csv("filterByRows.csv")
filterByColumns.to_csv("filterByColumns.csv")
filterRowsVSColumns.to_csv("filterRowsVSColumns.csv")
filterHEAD.to_csv("filterHEAD.csv")
filterTAIL.to_csv("filterTAIL.csv")
filterByComparison.to_csv("filterByComparison.csv")
filterAND.to_csv("filterAND.csv")
filterOR.to_csv("filterOR.csv")
filterNOT.to_csv("filterNOT.csv")
```
